```python
#copy all files from /content/Lexoid/examples/inputs to /content
import shutil
import os

# Source and destination paths
source_dir = "/content/Lexoid/examples/inputs"
destination_dir = "/content"

# Check if source directory exists
if not os.path.exists(source_dir):
    print(f"Source directory {source_dir} does not exist!")
else:
    # Copy all files from source to destination
    for file_name in os.listdir(source_dir):
        full_file_name = os.path.join(source_dir, file_name)
        if os.path.isfile(full_file_name):  # Ensure it's a file
            shutil.copy(full_file_name, destination_dir)
            print(f"Copied: {file_name}")

print("All files have been copied successfully!")
```

```python
import os
from lexoid.api import parse
from IPython.display import display, Image, Markdown
```

---
### Image Parsing - `LLM_PARSE`

```python
document_path = "/content/screenshot-1.png"
```

```python
Image(document_path, width=512)
```

```python
result_as_md = parse(document_path, "LLM_PARSE", raw=True)
```

```python
display(Markdown(result_as_md))
```
